```python



import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



df = pd.read_csv ("/Users/danort/Desktop/Ironhack labs/lab-customer-analysis-round-3/files_for_lab/csv_files/marketing_customer_analysis.csv")
df.head()
df.columns = df.columns.str.upper().str.replace(' ', '_').str.replace('[^a-zA-Z0-9_]', '')
df
```

```python
#Showing a plot of the total number of responses.
number_response = df['RESPONSE'].value_counts()
number_response.plot(kind='bar', color='lightblue', edgecolor='black')
plt.xlabel('Responses')
plt.ylabel('Count')
plt.title('Total Number of Responses')
```

```python
#plot of the response rate by the sales channel
sns.countplot(data=df, x="RESPONSE", hue="SALES_CHANNEL")
```

```python
# Showing a plot of the response rate by the total claim amountdf['RESPONSE'] = pd.to_numeric(df['RESPONSE'],errors='coerce')
df['RESPONSE'] = df['RESPONSE'].map({'yes': 1, 'no': 0})
df['claim_amount_range'] = pd.cut(df["TOTAL_CLAIM_AMOUNT"], bins=20)  # Adjust the number of bins as needed
response_rate_by_claim_amount = df.groupby('claim_amount_range')['RESPONSE'].mean()
sns.barplot(x="TOTAL_CLAIM_AMOUNT", y="claim_amount_range", data=df)
```

```python
df['RESPONSE'] = pd.to_numeric(df['RESPONSE'],errors='coerce')
df['CLAIM_AMOUNT_RANGE'] = pd.cut(df['TOTAL_CLAIM_AMOUNT'], bins=10)  
response_rate_by_claim_amount = df.groupby('CLAIM_AMOUNT_RANGE')['RESPONSE'].mean()
response_rate_by_claim_amount.hist(color='skyblue', edgecolor='black')
sns.distplot(df['TOTAL_CLAIM_AMOUNT'], bins=20)
plt.show()
```
